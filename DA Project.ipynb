```python
pygame.init()

class my_game:
    
    state=None   #coords array
    
   
    #define __inti__ variables
    
    def __init__ (self, running=False,reward=0):
        
        self.running=running
        
        self.reward=0
        
        self.clock=pygame.time.Clock()
        
        self.screen= pygame.display.set_mode([500, 500])
        
        self.c1=255
        
        self.c2=255
        
        self.rad=7
        
        self.color=(218, 112, 214)
        
        self.b=0
        
        self.a=[]
        
        self.change=False
        
        self.t= np.arange(-1.0, 1.0, 0.01)
        
        self.changepoint=1
        
        self.score=0
    
    def screen(self):
        
        return pygame.display.set_mode([500, 500])
    
    def generatecircle(self,c1,c2,color,radius,screen):
        
        cir =pygame.draw.circle(screen, color, (c1,c2), radius)
        
        return cir
    
    def quitgame(self):
        
        self.running=False
        
        return  pygame.quit()
         
    def fillscreen(self):
        
        return self.screen.fill((0,0,0))
        
    def moveup(self):
        
        return self.c2-2
    
    def moveright(self):
        
        return self.c1+2
    
    def moveleft(self):
        
        return self.c1-2
    
    def movedown(self):
        
        return self.c2+2
    
    def getcoordsfromrandomcircle(self,a,b):
        
        rdash=10
        self.p1=[]
        self.p2=[]
        for i in self.t:
            self.p1.append(int(rdash*np.cos(np.pi*i)+a))
            self.p2.append(int(rdash*np.sin(np.pi*i)+b))
        self.changepoint=0
        
        return [self.p1,self.p2,self.changepoint]
    
    def addcircle(self):
        
        if self.b<=2:
            
            self.a.insert(0,(self.c1,self.c2))
            #print(self.a, ": only inserting")
            
        else:

            if self.change:
                self.a.insert(0,(self.c1,self.c2))
                self.change=False
            else:
                self.a.insert(0,(self.c1,self.c2))
                self.a.pop(len(self.a)-1)

            #print(a)
        
    def checkcollision(self,p1,p2,c1,c2,rad,state):
        
        if state==1:
            if int(c1-rad) in p1:
                    if int(c2) in p2:
                        print("collision detected from right!")
                        return True
        elif state==2:
            if int(c1+rad) in p1:
                    if int(c2) in p2:
                        print("collision detected from left!")
                        return True
        elif state==3:
            if int(c1) in p1:
                    if int(c2-rad) in p2:
                        print("collision detected from bottom!")
                        return True
        elif state==4:
            if int(c1) in p1:
                    if int(c2-rad) in p2:
                        print("collision detected from top!")
                        return True
    
    def checkpupstate(self,c2,r2):
        
        if c2>r2:
            return 1
        else:
            return 0
        
    def checkpdownstate(self,c2,r2):
        
        if c2<r2:
            return 1
        else:
            return 0
    
    def checkprightstate(self,c1,r1):
        
        if r1>c1:
            return 1
        else:
            return 0
    
    def checkpleftstate(self,c1,r1):
        
        if r1<c1:
            return 1
        else:
            return 0
        
    def getactionmeaning(self):
        
        return ['up','right','down','left']
    
    def getobservationmeaning(self):
        
        return ['head coordinates', 'point coordinates', 'distance between them']
    
     
    def run_game(self , action=None):
        
        for event in pygame.event.get():

            if event.type == pygame.QUIT :

                self.quitgame()

            elif event.type == pygame.KEYDOWN:
                
                if event.key == pygame.K_p:
                    
                    self.reward+=1
                    
                    self.change=True
                
                if action==None:

                    if event.key == pygame.K_UP:

                        self.c2=self.moveup()

                        self.state=3

                    if event.key == pygame.K_RIGHT:

                        self.c1= self.moveright()

                        self.state=2

                    if event.key == pygame.K_DOWN:

                        self.c2= self.movedown()

                        self.state=4

                    if event.key == pygame.K_LEFT:

                        self.c1= self.moveleft()

                        self.state=1
                        
        if self.running:                
                        
                    
            if action!=None:

                if action==0:

                    self.c2=self.moveup()

                    self.state=3

                elif action==1:

                    self.c1= self.moveright()

                    self.state=2

                elif action==2:

                    self.c2= self.movedown()

                    self.state=4

                elif action==3:

                    self.c1= self.moveleft()

                    self.state=1


            #constant speed

            if self.state==3:

                self.c2=self.c2-5

            elif self.state==2:

                self.c1=self.c1+5

            elif self.state==4:

                self.c2=self.c2+5
                
            elif self.state==1:

                self.c1=self.c1-5



            self.addcircle() 

            self.b+=1

            if self.changepoint==1:

                self.f=random.randrange(20,480, 1)
                self.g=random.randrange(20,480, 1)

                self.y, self.z, self.changepoint= self.getcoordsfromrandomcircle(self.f,self.g)

                #print("coordinates aquired for point 1",self.y)


            self.fillscreen()       

            headcircle=self.generatecircle(self.c1,self.c2,self.color,self.rad,self.screen)

            ramdompoint=self.generatecircle(self.f,self.g,(255,0,0),self.rad,self.screen)


            for k in range(1,self.reward+1):

                if self.reward>=k:

                    self.generatecircle(self.a[k][0],self.a[k][1],(0,255,0),self.rad,self.screen)

            if self.checkcollision(self.y,self.z,self.c1,self.c2,self.rad,self.state):

                self.changepoint=1

                self.reward+=1

                self.score+=10

                self.change=True

            self.distance=math.sqrt((self.f-self.c1)**2+(self.g-self.c2)**2)

            if (self.distance<40):

                #print("snake close to the point")
                self.score+=1
                
            else:

                #print("snake away from the point")
                self.score-=1

            #checking state

            st1=self.checkpdownstate(self.c2,self.g)

            st2=self.checkpleftstate(self.c1,self.f)

            st3=self.checkprightstate(self.c1,self.f)

            st4=self.checkpupstate(self.c2,self.g)
            
            #print(self.reward)

            #print(self.a[0] ,self.a[1:])
            
            if self.reward>=3:
                if self.a[0] in self.a[1:]:
                        print("entered loop!")
                        print('the snake bit itself')
                        self.running=False
                        self.quitgame()
                        self.score-=100
                        pygame.quit()

                        return  [self.a[0][0], self.a[0][1],self.f,self.g,st1,st2,st3,st4,self.distance], self.score,not(self.running)

            if (self.c1>490 or self.c1<10 or self.c2>490 or self.c2<10):

                self.running=False
                self.quitgame()
                self.score-=100
                pygame.quit()
                return [self.a[0][0], self.a[0][1],self.f,self.g,st1,st2,st3,st4,self.distance], self.score,not(self.running)


            self.clock.tick(20)

            pygame.display.update()

            return [self.a[0][0], self.a[0][1],self.f,self.g,st1,st2,st3,st4,self.distance], self.score,not(self.running)
```

# 2. Building DQNAgent

### Import dependencies

```python
from collections import deque
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import os
```

### Set parameters based on our environment

```python
state_size=9
action_size=4
batch_size=32
no_of_episodes= 1001         #we will randomly remember something that happened in the previous episodes
```

```python
model_save_dir='model_ouput/snake_game_model'    #directory to save modeloutput

if not os.path.exists(model_save_dir):
    os.makedirs(model_save_dir)
```

### Defining Agent

```python
class DQNAgent:
    
    def __init__(self, state_size, action_size):
        self.state_size=state_size
        self.action_size=action_size
        
        self.memory= deque(maxlen=2000)   #memory replay deque
        
        #Hyperparametrs
        
        self.gamma= 0.95   #discounting parameter for discounting future rewards
        
        self.epsilon=1.0  #for action change from eploration to eploitation
        
        self.epsilon_decay=0.995  #decreasing epsilon value to slowly shift from random env exploration to exploit what our agent learns
        
        self.epsilon_min= 0.001  #min epsilon for exploration
        
        self.learning_rate=0.001  # for gradient descent
        
        #building Dense NN model for approximation of q* over a given state 
        
        
        #defining build model method
        self.model= self._build_model()
        
    def _build_model(self):

        model= Sequential()

        model.add(Dense(24, input_dim= self.state_size, activation='relu'))
        model.add(Dense(32,activation='relu'))
        model.add(Dense(self.action_size,activation='linear')) #we donot want prob we want abstract acion as our outputs we we use linear activation
        model.compile(loss='mse', optimizer=Adam(learning_rate=self.learning_rate))

        return model

    def get_model_sumary(self):

        return self.model.summary()

    def remember(self, state, action, reward, next_state, done):  # appending attr to memory deque

        self.memory.append(( state, action, reward, next_state, done))

    def act(self, state):   # we will try to figure out what action to take given a state

        #randomly explore action or exploit action based on prevoious learning

        if np.random.rand() <= self.epsilon:  # if random value b/w 0 and 1 is <= epsilon then we will act randomly

            return random.randrange(self.action_size)  # random action to observe env.

        #overtime epsilon will decay and we will shift towards exploitation

        act_values = self.model.predict(state) #prediction by our NN model

        return np.argmax(act_values[0])

    def replay(self, batch_size):  # **

        minibatch= random.sample(self.memory, batch_size)  #randomly sampling from our memory deque of our batch_size=32

        for state, action, reward, next_state, done in minibatch:   #iteration through our memory minibatch
            target= reward  # game is done

            if not done:

                # discounted future reward

                target= (reward+ self.gamma* np.amax(self.model.predict(next_state)))

                #map max future reward to current reward

                target_f= self.model.predict(state)
                target_f[0][action]= target

                #fit our model to train

                self.model.fit(state,target_f, epochs=1,verbose=1)

            #decresing epsilon for the shift

            if self.epsilon>self.epsilon_min:
                self.epsilon*=self.epsilon_decay

    def loadmodel(self,name):

        #loading model from dir

        self.model.load_weights(name)

    def save(self,name):

        #saving model to dir

        self.model.save_weights(name)
```

### Defining agent

```python
agent= DQNAgent(state_size, action_size)
```

## Making agent act with the environment

### Training the agent:

```python
for episode in range(no_of_episodes):
    game=my_game(True)
    
    state,rew,done= game.run_game()  #running our game
        
    state=np.reshape(state,[1,state_size])
    
    for time in range(5000):
        
        action=agent.act(state)
        
        next_state, rew,done = game.run_game(action)
        
        next_state=np.reshape(next_state,[1,state_size])
        
        agent.remember( state, action, rew, next_state, done)
        
        state=next_state

        if done:
            break
        #giving time to update weights
            
    if (len(agent.memory) > batch_size):
        agent.replay(batch_size)
    
    if episode%50 ==0:
        agent.save(model_save_dir+ "weights_" + "{:04d}" .format(episode) +".hdf5")
            
    x='-'
    print("episode: {} ,Score: {} , Epsilon : {:.2}".format(episode, rew, agent.epsilon))
    print(x*45)
```
